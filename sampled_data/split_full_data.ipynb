```python
import numpy as np
import pandas as pd 

path_rating = 'ratings.csv'
path_links = 'links.csv'
```

```python
df_rating = pd.read_csv(path_rating)
df_rating[['userId', 'movieId', 'rating']] = df_rating[['userId', 'movieId', 'rating']].astype(int)
print(len(df_rating))
print(df_rating['movieId'].nunique(), df_rating['movieId'].max())
df_rating.head()
rating_with_nan = df_rating[df_rating.isna().any(axis=1)]
rating_with_nan.head()
```

```python
user_rating_counts = df_rating['userId'].value_counts()

# Lọc ra các userId có số lần rating ít hơn 10
filtered_userIds = user_rating_counts[user_rating_counts >= 10].index

# Lọc DataFrame để chỉ giữ lại các hàng có userId thuộc danh sách filtered_userIds
filtered_df1 = df_rating[df_rating['userId'].isin(filtered_userIds)]
filtered_df1
```

```python
std_by_user = filtered_df1.groupby('userId')['rating'].std()
print(std_by_user)
print(std_by_user[std_by_user > 0.5])
filtered_users = std_by_user[std_by_user > 0.5].index

filtered_df2 =  filtered_df1[filtered_df1['userId'].isin(filtered_users)]
filtered_df2
```

```python
df_links = pd.read_csv(path_links)
print(len(df_links))
print(df_links['movieId'].nunique(), df_links['movieId'].max())
print(df_links['imdbId'].nunique())
print(df_links['tmdbId'].nunique())
df_links.head()
```

```python
df_link_ratings = pd.merge(df_links, df_rating, on='movieId', how='inner') 
print(len(df_link_ratings))
df_link_ratings.head()
```

```python
rows_with_nan_0 = df_links[df_links['movieId'].isna()]
print('nan movieId:\n', rows_with_nan_0)
print()
rows_with_nan_1 = df_links[df_links['imdbId'].isna()]
print('nan imdbId:\n', rows_with_nan_1)
print()
rows_with_nan_2 = df_links[df_links['tmdbId'].isna()]
print('nan tmdbId:\n', rows_with_nan_2)
```

```python
df_link_ratings = df_link_ratings.dropna()
```
